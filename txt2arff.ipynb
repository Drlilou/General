```
file="Groceries.txt"
```

```
f = open(file, "r")
legume=[]
data=[]
for i in f:
  s=i.split("\\")[1]
  s=s.replace('\n','')
  s=s.split(",")
  data.append(s)
  legume.extend(s)
```

```
l=list(set(legume))
l.sort()
```

```
newData=[]
for i in data:
  datarow=[]
  for  food in l :
     if  food in i:
       datarow.append(1)
     else:
       datarow.append(0)
  newData.append(datarow)

```

```
import pandas as pd
```

```
df=pd.DataFrame(newData)
df.columns =l
df.head()
```
